# Vizualizing Inconel Summaries

# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

# Creating list of Inconel's prediction summaries

```python
csv = glob.glob("*Summary.xlsx")
```

```python
csv
```

# Creating and Combining all Inconel summaries into a DataFrame

```python
# Creating empty DataFrame
df = pd.DataFrame()

# Iterating list of csv names
for i in csv:
    
    # Reading csv and saving to variable a
    a = pd.read_excel(i, usecols = ["SR", "MRR"])
    
    # Separating grade name of Inconel
    grade_name = i.replace("_Summary.xlsx", "")
    
    # Creating a column of inconel grade name
    a["Grade"] = grade_name
    
    # Appending each Inconel grade's SR and MRR predictions in DataFrame df
    df = pd.concat([df,a])
```

# Check if all the Grades are in the df

```python
df.Grade.unique()
```

# Display first 10 rows of the DataFrame df

```python
df.head(10)
```

# Plotting SR vs Inconel Grades

```python
sns.set(style="ticks")
sns.set_style("darkgrid")

s = sns.catplot(data = df, x = "SR", y = "Grade" ,hue = "Grade", aspect= 2, height = 6).set(
    title = 'Type vs SR (Predictions)')

s.savefig("Type_vs_SR_(Predictions).jpg")
```

# Plotting MRR vs Inconel Grades

```python
s = sns.catplot(data = df, x = "MRR", y = "Grade" ,hue = "Grade", aspect= 2, height = 6).set(
    title = 'Type vs MRR (Predictions)')
plt.xscale('log')

s.savefig("Type_vs_MRR_(Predictions).jpg")
```
